## Functions

```python
def date_time_columns(df, month, year):
  """
  Function that transforms timestamp to Date/Time
  and creates new columns for Year, Month, Day, Day of Week, Hour
  """
  # creates new column with Date/Time, delete Timestamp column and rename
  df['last_reported_2'] = pd.to_datetime(df['last_reported'], unit='s')
  df = df.drop(['last_reported'], axis=1)
  df.rename(columns={"last_reported_2": "last_reported"}, inplace=True)

  #create new columns
  df['date'] = pd.DatetimeIndex(df['last_reported']).date
  df['year'] = pd.DatetimeIndex(df['last_reported']).year
  df['month'] = pd.DatetimeIndex(df['last_reported']).month
  df['day'] = pd.DatetimeIndex(df['last_reported']).day
  df['day_of_week'] = pd.DatetimeIndex(df['last_reported']).dayofweek
  df['hour'] = pd.DatetimeIndex(df['last_reported']).hour

  df = df[(df['year'] == year) & (df['month'] == month)]

  return df
```

```python
def dock_availability_normalized(df, columns_groupby):
  """
  Function that creates column for percentage of dock availability
  and group dates by available mean
  """
  # creates column 'perc_available'
  df['perc_available'] = (df['num_docks_available']/df['capacity'])
  df = df[(df['perc_available'] >= 0) & (df['perc_available'] <= 1)]

  #group by mean
  df_gb = df.groupby(columns_groupby, as_index=False)['perc_available'].mean()
  df_gb['perc_available'] = df_gb['perc_available'].round(2)
  return df_gb
```

```python
def check_month(df, month):
  """
  Function that checks if month data is accurate,
  reset index and sort values
  """
  df_clean = df[df.month == month]
  df_clean.reset_index(drop=True)

  #sort by hour
  df_clean.sort_values(by=['station_id', 'month', 'day', 'hour'], inplace=True)

  return df_clean
```

```python
def contexts_5h(df):
    """
    Function that creates contexts for every 5h
    """
    df_context = df.iloc[4::5, :].reset_index(drop=True)

    df_context['ctx-4'] = df['perc_available'].iloc[0::5].reset_index(drop=True)
    df_context['ctx-3'] = df['perc_available'].iloc[1::5].reset_index(drop=True)
    df_context['ctx-2'] = df['perc_available'].iloc[2::5].reset_index(drop=True)
    df_context['ctx-1'] = df['perc_available'].iloc[3::5].reset_index(drop=True)

    return df_context
```

```python
def df_month(df):
  """
  Function that removes NAN values
  """
  df = df.dropna()
  df = df.reset_index(drop=True)

  return df
```

## Preparing and creating complete data frame:
For each month and year from web with station information:

```python
#rewrite months cause it saves Març from web
i2m = list(zip(range(1,13), ['Gener','Febrer','Març','Abril','Maig','Juny','Juliol','Agost','Setembre','Octubre','Novembre','Desembre']))

#variables has the columns of interest **can be changed later**
variables = ['station_id','num_docks_available','status','last_reported']

#create a final DF to be used in the model
df_final = pd.DataFrame()

# Loop over all months and years
for year in [2022, 2021, 2020, 2019]:

    if year==2019:
      #remove cause 2019 has no data for months: 01 and 02
      i2m = i2m[2:]

    for month, month_name in i2m:

        print(f"{year}_{month:02d}_{month_name}_BicingNou_ESTACIONS.csv")

        print('Preparing data frame...')

        df_read_csv = pd.read_csv(f"{year}_{month:02d}_{month_name}_BicingNou_ESTACIONS.csv")

        #get reduced status data frame
        df_bicing_reduced = df_read_csv.filter(variables, axis=1)
        df_bicing_reduced = df_bicing_reduced[df_bicing_reduced.status == 'IN_SERVICE']

        #merge bicing DF with station info DF
        df_merged = pd.merge(df_bicing_reduced, df_station_reduced, on='station_id', how='inner')

        #date time transformation and new columns
        df_merged = date_time_columns(df_merged, month=month, year=year)

        #compute dock availabity and groups DF
        columns_to_keep = ['station_id', 'lat', 'lon', 'altitude', 'address', 'post_code', 'capacity', 'date', 'year',
                          'month', 'day', 'day_of_week', 'hour']
        df_grouped = dock_availability_normalized(df_merged, columns_groupby=columns_to_keep)

        #check if has only current month on data set and sort by hour
        df_clean = check_month(df_grouped, month)

        #create contexts
        df_contexts = contexts_5h(df_clean)

        #get df of current month
        df_sample = df_month(df_contexts)

        print('Data frame of ' + str(year) + ' in month ' + str(month) + ' was created')

        #append to final df and create complete DF
        df_final = pd.concat([df_final, df_sample])

        print('Data frame appended to final df')

        print('------------------------------------------------------------------------------------')
```
