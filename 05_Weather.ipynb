## Weather:

```python
csv_weather = pd.read_csv('/content/drive/My Drive/CapstoneProject_NathyDani/api_weather_opendata.csv', sep=';')
csv_weather
```

```python
df_weather = csv_weather.loc[:, 'fecha':'velmedia']
df_weather.head()
```

```python
df_weather = df_weather.drop(columns=['provincia', 'tmin', 'horatmin', 'tmax', 'horatmax', 'dir'])
df_weather.head()
```

```python
df_weather = df_weather.rename(columns={'fecha': 'date', 'nombre': 'name', 'tmed': 'temp', 'velmedia': 'wind'})
df_weather.head()
```

```python
print(df_weather['indicativo'].unique())
print(df_weather['name'].unique())
print(df_weather['altitud'].unique())
```

```python
df_weather = df_weather.drop(columns=['indicativo', 'name', 'altitud'])
df_weather.head()
```

```python
df_weather.isnull().sum()
```

```python
df_weather.info()
```

```python
df_weather['prec'] = df_weather['prec'].astype(float)
```

```python
df_weather['prec'].unique()
```

```python
df_weather['prec'] = df_weather['prec'].replace('Ip', '')
df_weather = df_weather.replace(r'^\s*$', np.nan, regex=True)
```

```python
df_weather['prec'] = df_weather['prec'].astype(float)
```

```python
df_weather[df_weather.isna().any(axis=1)]
```

```python
df_weather = df_weather.interpolate(method='linear', limit_direction='forward', axis=0)
```

```python
df_weather[df_weather.isna().any(axis=1)]
```

```python
df_weather.isnull().sum()
```

```python
df_weather.describe()
```

```python
df_weather['date'] = df_weather['date'].astype('datetime64[ns]')
```
